# FFT w/ Signal Reconstruction

This breaks down a series of instrument samples using fft and then utilizes the frequency and magnitudes to 
reconstruct the original signal. Timbre is remarkably accurate despite the absence of an envelope.

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import os
import IPython.display as ipd
```

```python
BASE_DIR = 'audio/'

viola_path = 'viola.wav'
piano_path = 'piano.wav'
sax_path = 'sax.wav'
chime_path = 'chime.wav'
```

```python
viola, sr = librosa.load(os.path.join(BASE_DIR, viola_path))
piano, sr = librosa.load(os.path.join(BASE_DIR, piano_path))
sax, sr = librosa.load(os.path.join(BASE_DIR, sax_path))
chime, sr = librosa.load(os.path.join(BASE_DIR, chime_path))
```

```python
def harmonic_logic(h_array, h_index, m_index, ms):
    check_index = h_index - 1
    
    if( h_array.size == 0):
        h_array = np.append(h_array, int(m_index) )
    elif ( m_index - h_array[check_index] < 100 ):
        m = ms[m_index]
        mag_index = int(h_array[check_index])
        current = ms[mag_index]
        if m == current:
            pass
        max_index = np.argmax([m, current])
        h_array[check_index] = int(m_index) if max_index == 0 else h_array[check_index]
    elif ( m_index - h_array[h_index - 1] > 100 ):
        h_array = np.append(h_array, int(m_index) )
    
    return h_array

def frequency_array(ms, freq, bins):
    window = 200
    hop = 50
    harmonics = np.array([])
    harmonic_index = 0
    
    for i in range(bins):
        s = harmonics.size
        max_magnitude = i + np.argmax(ms[i:(i+window)])
        harmonics = harmonic_logic(harmonics, harmonic_index, max_magnitude, ms)
        i = i + hop
        harmonic_index = harmonic_index + 1 if s < harmonics.size else harmonic_index
        
    return harmonics
    
          
```

```python
def convert_from_indices_to_values(mag_array, freq_array, indices):
    values = []
    for i in indices:
        index = int(i)
        value = [mag_array[index], freq_array[index]]
        values.append(value)
        
    return values
```

```python
def define_mag_freq_arrays(signal, sr, f_ratio=1):
    ft = np.fft.fft(signal)
    magnitude_spectrum = np.abs(ft)
    frequency = np.linspace(0, sr, len(magnitude_spectrum))
    num_freq_bins = int(len(frequency) * f_ratio)
    return [magnitude_spectrum, frequency, num_freq_bins]

def retrieve_highest_values(value_dict):
    mag_array, freq_array, num_bins = value_dict
    bins = frequency_array(mag_array, freq_array, num_bins)
    return convert_from_indices_to_values(mag_array, freq_array, bins)
    

def plot_magnitude_spectrum(value_dict, title):
    mag_array, freq_array, num_bins = value_dict
    #plot spectrum
    plt.figure(figsize=[18,5])
    plt.plot(freq_array[:num_bins], mag_array[:num_bins])
    plt.xlabel("Frequency (Hz)")
    plt.title(title)
    
    plt.show()
```

```python
import pyaudio
import math

p = pyaudio.PyAudio()
duration = 2.0
```

```python
def makesine(freq, dur):
    t = np.linspace(0, dur, math.ceil(sr*dur))
    x = np.sin(2 * np.pi * freq * t)
    return x



def sine_wave_iterator(values):
    max_magnitude = max(max(values, key=lambda x: x[0]))
    t = np.linspace(0, duration, math.ceil(sr*duration))
    out = np.zeros(t.size)
    for value in values:
        magnitude = value[0]
        frequency = value[1]
        amp = (magnitude / max_magnitude) / len(values)
        sine = makesine(frequency, duration)
        x = sine * amp
        out = out + x
    if np.max(out)>abs(np.min(out)):
        out = out / np.max(out)
    else:
        out = out / -np.min(out)
    return out

def render_reconstructed_waveforms(values):
    t = np.linspace(0, duration, math.ceil(sr*duration))
    waveform = sine_wave_iterator( values )
    plt.plot(t, waveform)
    plt.xlim(0, 0.005)
    return waveform
```

```python
ipd.Audio(os.path.join(BASE_DIR, viola_path))
```

```python
value_dict = define_mag_freq_arrays(viola, sr, 0.5)
plot_magnitude_spectrum(value_dict, "violin")

values = retrieve_highest_values(value_dict)

waveform = render_reconstructed_waveforms(values)
ipd.Audio(waveform, rate=sr)
```
